### Imports

```python
import susi
import sompy
from time import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
from scipy.cluster.hierarchy import dendrogram
```

### Carregando dados

```python
colunas = ['Age', 'Gender', 'CP', 'Trestbps', 'Chol', 'FBS', 'RestECG', 'Thalach', 'Exang', 'Oldpeak', 'Slope', 'CA', 'Thal', 'Goal']
cleve = pd.read_csv('./heartdisease/processed_cleveland.data', na_values = '?')
cleve.columns = colunas
hung  = pd.read_csv('./heartdisease/processed_hungarian.data', na_values = '?')
hung.columns = colunas
swiss  = pd.read_csv('./heartdisease/processed_switzerland.data', na_values = '?')
swiss.columns = colunas
va  = pd.read_csv('./heartdisease/processed_va.data', na_values = '?')
va.columns = colunas
heart = pd.concat([cleve, hung, swiss, va])
heart.loc[heart['Goal']> 0, 'Goal'] = 1
heart.drop('CA', axis='columns', inplace = True)
heart.dropna(subset=['Trestbps', 'Thalach', 'Exang', 'Oldpeak', 'Chol', 'RestECG', 'FBS'], inplace = True)
heart.loc[heart['Thal'].isna(), 'Thal'] = 0
heart.loc[heart['Slope'].isna(), 'Slope'] = 0
dados_heart = heart.iloc[:,:-1].values
label_heart = heart.iloc[:,-1].values

scaler_heart = MinMaxScaler()
scaler_heart.fit(dados_heart)
dados_heart = scaler_heart.transform(dados_heart)
```

```python
pima = pd.read_csv('diabetes.csv',delimiter = ',')

# Carregando dados
dados_pima = pima.iloc[:,:-1].values
label_pima = pima.iloc[:,-1].values

scaler_pima = MinMaxScaler()
scaler_pima.fit(dados_pima)
dados_pima = scaler_pima.transform(dados_pima)
```

### Definindo funções

```python
def find_optimal_clusters(data, max_k):
    iters = range(2, max_k+1, 2)
    
    sse = []
    for k in iters:
        sse.append(KMeans(n_clusters=k, random_state=20).fit(data).inertia_)
        print('Agrupamento {} treinado'.format(k))
        
    f, ax = plt.subplots(1, 1)
    f.set_figheight(8)
    f.set_figwidth(15)
    ax.plot(iters, sse, marker='o')
    ax.set_xlabel('Cluster Centers')
    ax.set_xticks(iters)
    ax.set_xticklabels(iters)
    ax.set_ylabel('SSE')
    ax.set_title('SSE by Cluster Center Plot')
```

```python
def plot_pca3d(X, y):    
    fig = plt.figure(1, figsize=(15, 8))
    ax = Axes3D(fig, rect=[-1, -1, 1, 1], elev=30, azim=120)

    plt.cla()
    pca = PCA(n_components=3)
    pca.fit(X)
    X = pca.transform(X)
    
    for name, label in [('Sem', 0), ('Com', 1)]:
        ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean(),
              X[y == label, 2].mean(),
              name, horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))

    y = np.choose(y, [0, 1]).astype(np.float)
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral,
               edgecolor='k')
```

```python
def plot_pca2d(X, labels):
    
    reduced_data = PCA(n_components=2).fit_transform(X)
    kmeans = KMeans(init='k-means++', n_clusters=2, n_init=10)
    kmeans.fit(reduced_data)

    h = .02 

    # Plot the decision boundary. For that, we will assign a color to each
    x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
    y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Obtain labels for each point in mesh. Use last trained model.
    Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure(1, figsize = (15,8))
    plt.clf()
    plt.imshow(Z, interpolation='nearest',
               extent=(xx.min(), xx.max(), yy.min(), yy.max()),
               cmap=plt.cm.Paired,
               aspect='auto', origin='lower')

    plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=15)
    # Plot the centroids as a white X
    centroids = kmeans.cluster_centers_
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='x', s=250, linewidths=3,
                color='w', zorder=10)
    plt.title('Agrupamento K-means com PCA\n')
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.xticks(())
    plt.yticks(())
    plt.show()
    bench_k_means(kmeans, 'kmeans pca', X, labels)
```

```python
def bench_k_means(estimator, name, data, labels):
    t0 = time()
    estimator.fit(data)
    print('Nome\t\tTempo\tInercia   Homogenidade\tPlenitude\tHarmonico HP\tSilhueta')
    print('%-9s\t%.2fs\t%i\t\t%.3f\t%.3f\t\t%.3f\t\t%.3f'
          % (name, (time() - t0), estimator.inertia_,
             metrics.homogeneity_score(labels, estimator.labels_),
             metrics.completeness_score(labels, estimator.labels_),
             metrics.v_measure_score(labels, estimator.labels_),
             metrics.silhouette_score(data, estimator.labels_,
                                      metric='euclidean')))
```

```python
def plot_dendrogram(model, **kwargs):
    # Cria a matriz de ligação e plota dendrograma

    # cria as amostras de contagem para cada nó
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # plota o dendrograma
    dendrogram(linkage_matrix, **kwargs)
```

```python
kmeans_heart = KMeans(n_clusters=2, init='k-means++', max_iter=100).fit(dados_heart)
```

```python
bench_k_means(kmeans_heart, name="k-means heart", data=dados_heart, labels = label_heart)
```

```python
plot_pca3d(dados_heart, label_heart)
```

```python
plot_pca2d(dados_heart, label_heart)
```

```python
find_optimal_clusters(dados_heart, 30)
```

```python
kmeans_heart = KMeans(n_clusters=6, init='k-means++', max_iter=100)
kmeans_heart.fit(dados_heart)
```

```python
goal_0 = []
goal_1 = []
for i in range(label_heart.shape[0]):
    if label_heart[i] == 0:
        goal_0.append(kmeans_heart.labels_[i])
    elif label_heart[i] == 1:
        goal_1.append(kmeans_heart.labels_[i])

goal_0 = np.array(goal_0)
goal_1 = np.array(goal_1)
```

```python
unique0, counts0 = np.unique(goal_0, return_counts=True)
unique1, counts1 = np.unique(goal_1, return_counts=True)

fig, axs = plt.subplots(1, 2, figsize=(15,6), sharey=True, tight_layout=True)
axs[0].plot(unique0, counts0)
axs[1].plot(unique1, counts1)
```

```python
# configurnado distance_threshold=0 garante que será calculado uma árvore completa.
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
model = model.fit(dados_heart)
```

```python
plt.figure(figsize=(15,8))
plt.title('Hierarchical Clustering Dendrogram')
# plot the top three levels of the dendrogram
plot_dendrogram(model, truncate_mode='level', p=3)
plt.xlabel("Número de pontos no nó (ou index do ponto se não tiver parêntese).")
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=3)
model = model.fit(dados_heart)
```

```python
print('Nome\t\t   Homogenidade\tPlenitude\tHarmonico HP\tSilhueta')
print('%-9s\t\t%.3f\t%.3f\t\t%.3f\t\t%.3f'
      % ('hierarchical', metrics.homogeneity_score(label_heart, model.labels_),
         metrics.completeness_score(label_heart, model.labels_),
         metrics.v_measure_score(label_heart, model.labels_),
         metrics.silhouette_score(dados_heart, model.labels_,
                                  metric='euclidean')))
```

```python
goal_0 = []
goal_1 = []
goal_2 = []
for i in range(label_heart.shape[0]):
    if label_heart[i] == 0:
        goal_0.append(model.labels_[i])
    elif label_heart[i] == 1:
        goal_1.append(model.labels_[i])

goal_0 = np.array(goal_0)
goal_1 = np.array(goal_1)
```

```python
unique0, counts0 = np.unique(goal_0, return_counts=True)
unique1, counts1 = np.unique(goal_1, return_counts=True)

fig, axs = plt.subplots(1, 2, figsize=(15,6), sharey=True, tight_layout=True)
axs[0].plot(unique0, counts0)
axs[1].plot(unique1, counts1)
```

```python
mapsize = [10,10]
som = sompy.SOMFactory.build(dados_heart, mapsize, mask=None, mapshape='planar', lattice='rect', normalization='var', initialization='pca', neighborhood='gaussian', training='batch', name='sompy')  # this will use the default parameters, but i can change the initialization and neighborhood methods
som.train(n_job=1, verbose='info')  # verbose='debug' will print more, and verbose=None wont print anything
```

```python
v = sompy.mapview.View2DPacked(20, 20, title = 'Pima',text_size=10)
v.show(som, what='codebook', which_dim='all', col_sz=4)
```

```python
v = sompy.mapview.View2DPacked(2, 2, 'pima',text_size=8)  
cl = som.cluster(n_clusters=2)
v.show(som, what='cluster')
```

```python
plt.clf()
h = sompy.hitmap.HitMapView(10, 10, 'hitmap', text_size=8, show_text=True)
_ =h.show(som)
```

```python
u = sompy.umatrix.UMatrixView(50, 50, 'umatrix', show_axis=True, text_size=8, show_text=True)

#This is the Umat value
UMAT  = u.build_u_matrix(som, distance=1, row_normalized=False)

#Here you have Umatrix plus its render
UMAT = u.show(som, distance2=1, row_normalized=False, show_data=True, contooor=True, blob=False)
```

```python
som = susi.SOMClustering(n_rows=10, n_columns=10)
som.fit(dados_heart)
```

```python
u_matrix = som.get_u_matrix()
susi.SOMPlots.plot_umatrix(u_matrix, 10, 10)
plt.show()
```

```python
clusters = som.get_clusters(dados_heart)
plt.scatter(x=[c[1] for c in clusters], y=[c[0] for c in clusters], c=label_heart, alpha=0.2)
plt.gca().invert_yaxis()
plt.show()
```

```python
susi.SOMPlots.plot_som_histogram(som.get_bmus(dados_heart), 10,10)
```
