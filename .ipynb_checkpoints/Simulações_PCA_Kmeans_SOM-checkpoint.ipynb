# Definindo Algumas FunÃ§Ãµes

```python
def find_optimal_clusters(data, max_k):
    iters = range(2, max_k+1, 2)
    
    sse = []
    for k in iters:
        sse.append(KMeans(n_clusters=k, random_state=20).fit(data).inertia_)
        print('Agrupamento {} treinado'.format(k))
        
    f, ax = plt.subplots(1, 1)
    f.set_figheight(8)
    f.set_figwidth(15)
    ax.plot(iters, sse, marker='o')
    ax.set_xlabel('Cluster Centers')
    ax.set_xticks(iters)
    ax.set_xticklabels(iters)
    ax.set_ylabel('SSE')
    ax.set_title('SSE by Cluster Center Plot')
```

```python
def bench_k_means(estimator, name, data, labels):
    estimator.fit(data)
    print(name)
    print('Homogeneity: ', metrics.homogeneity_score(labels, estimator.labels_))
    print('Completeness: ', metrics.completeness_score(labels, estimator.labels_))
    print('V-measure: ', metrics.v_measure_score(labels, estimator.labels_))
    print('Silhueta: ', metrics.silhouette_score(data, estimator.labels_, metric='euclidean'))
```

```python
# utiliza o PCA para reduÃ§Ã£o da dimensionalidade dos dados
def plot_pca3d(X, y):    
    fig = plt.figure(1, figsize=(15, 8))
    ax = Axes3D(fig, rect=[-1, -1, 1, 1], elev=30, azim=120)

    plt.cla()
    pca = PCA(n_components=3)
    pca.fit(X)
    X = pca.transform(X)
    
    for name, label in [('Sem', 0), ('Com', 1)]:
        ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean(),
              X[y == label, 2].mean(),
              name, horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))

    y = np.choose(y, [0, 1]).astype(np.float)
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral,
               edgecolor='k')
```

```python
def plot_dendrogram(model, **kwargs):
    # Cria a matriz de ligaÃ§Ã£o e plota dendrograma

    # cria as amostras de contagem para cada nÃ³
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # plota o dendrograma
    dendrogram(linkage_matrix, **kwargs)
```

# MÃ©tricas de AvaliaÃ§Ã£o

MÃ©tricas Usadas:

* homogeneity_score -> MÃ©trica de homogeneidade de uma rotulagem de cluster dada uma verdade bÃ¡sica. Um resultado de agrupamento satisfaz a homogeneidade se todos os clusters contiverem apenas pontos de dados que pertencem a uma Ãºnica classe.

* Completeness_score -> MÃ©trica de integridade de uma rotulagem de cluster dada uma verdade bÃ¡sica. Um resultado de armazenamento em cluster satisfaz a integridade se todos os pontos de dados que sÃ£o membros de uma determinada classe sÃ£o elementos do mesmo cluster.

* V_measure_score -> A medida V Ã© a mÃ©dia harmÃ´nica entre homogeneidade e integridade:

    v = (1 + beta) * completeness * homogeneity / (beta * homogeneity + completeness)

* Silhouette -> O Coeficiente Silhouette Ã© calculado usando a distÃ¢ncia mÃ©dia intra-cluster ( a) e a distÃ¢ncia mÃ©dia do cluster mais prÃ³ximo ( b) para cada amostra.

# Diabetes

Utilizando KMeans para treinamento em atÃ© 2 subconjuntos

Utilizando metricas de avaliaÃ§Ã£o de modelos com o Sklearn

```python
kmeans_diabetes = KMeans(n_clusters=2, init='k-means++', max_iter=100).fit(diabetes_data)
bench_k_means(kmeans_diabetes, 'Kmeans Diabetes', diabetes_data, diabetes_labels)
```

Utilizando o KMeans para treinamento em atÃ© 30 subconjuntos

```python
find_optimal_clusters(diabetes_data, 30)
```

```python
plot_pca3d(diabetes_data, diabetes_labels)
```

Utilizando o AgglomerationClustering na base de Dados de Diabetes

```python
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
model = model.fit(diabetes_data)
```

```python
plt.figure(figsize=(15,8))
plt.title('Hierarchical Clustering Dendrogram')
# plot the top three levels of the dendrogram
plot_dendrogram(model, truncate_mode='level', p=3)
plt.xlabel("NÃºmero de pontos no nÃ³ (ou index do ponto se nÃ£o tiver parÃªntese).")
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=2)
model = model.fit(diabetes_data)
```

```python
bench_k_means(model, 'Agglomeration Diabetes', diabetes_data, diabetes_labels)
```

Utilizando o SOM para visualizaÃ§Ã£o de dados (Mapa)

```python
mapsize = [10,10]
som = sompy.SOMFactory.build(diabetes_data, mapsize, mask=None, mapshape='planar', lattice='rect', normalization='var', initialization='pca', neighborhood='gaussian', training='batch', name='sompy')
som.train(n_job=1, verbose='info') 
```

```python
v = sompy.mapview.View2DPacked(20, 20, title = 'Diabetes',text_size=10)
v.show(som, what='codebook', which_dim='all', col_sz=4)
```

```python
plt.clf()
h = sompy.hitmap.HitMapView(10, 10, 'hitmap', text_size=8, show_text=True)
h.show(som)
```

```python
som = susi.SOMClustering(n_rows=10, n_columns=10)
som.fit(diabetes_data)
```

```python
matrix = som.get_u_matrix()
susi.SOMPlots.plot_umatrix(matrix, 10, 10)
plt.show()
```

```python
clusters = som.get_clusters(diabetes_data)
plt.scatter(x=[c[1] for c in clusters], y=[c[0] for c in clusters], c=diabetes_labels, alpha=0.2)
plt.gca().invert_yaxis()
plt.show()
```
